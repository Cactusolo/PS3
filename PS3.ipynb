## Problem 3 (4 pts):
Write a script (or Jupyter Notebook code block) that calculates the fastest rise in water level per 6-minute period between measurements and reports the data and time that was observed and the change in water level during that period. You can use line by line or dataframe for this.

```python
import pandas as pd

df=pd.read_csv("./CO-OPS__8729108__wl.csv")

#remove NA
data=df.dropna()

#extract the data frame called "data" that only
#containing two columes --- 'Date Time' and ' Water Level'
data=pd.DataFrame(data, columns=['Date Time', ' Water Level'])
data[' Water Level'] = pd.to_numeric(data[' Water Level'], errors='coerce')
#data[1:722]
#n=0
raise_max=0.0
for i in range(0,data.shape[0]-1): #loop through the data frame based on its dimension
    #n=n+1
    level_diff=float(data.iloc[i+1][1])-float(data.iloc[i][1]) #caculate level difference
    #print(n, level_diff)
    if level_diff > raise_max:
        raise_max=level_diff #maxi raised level
        start=data.iloc[i][0] #starting date
        end=data.iloc[i+1][0] #ending date
print("From", start, "to", end, "water levle increased:", round(raise_max, 2))
            
```

## Extra Credit (4 pts):
Using the Web Services Button on [this page](https://tidesandcurrents.noaa.gov/waterlevels.html?id=8729108&units=standard&bdate=20181011&edate=20181012&timezone=GMT&datum=MLLW&interval=6&action=data) as a guide and/or the [API description](https://tidesandcurrents.noaa.gov/api/), write a script (or Jupyter Notebook code block) to download data from Panama City, Fl for "today" e.i. whenever the code is run, get that day's data.

```python
import urllib.request, urllib.parse, urllib.error
import datetime as dt
import requests

today=dt.datetime.today().strftime('%Y%m%d')

serviceurl = 'https://tidesandcurrents.noaa.gov/api/'
#id="8729108"
#units="standard"
bdate=today
edate=today
#timezone="GMT"
#datum="MLLW"
#interval=6
#action=data
output=open("water_levle_today.csv", "w+")
while True:
    url = serviceurl + "datagetter?product=water_level&application=NOS.COOPS.TAC.WL&" + urllib.parse.urlencode(
        {'begin_date': bdate}) + "&" + urllib.parse.urlencode({'end_date': edate}) + "&datum=MLLW&station=8729108&time_zone=gmt&units=english&format=csv" 

    #print('Retrieving', url)
    response = requests.get(url)
    #requests.get(url)
if response.status_code != 200:
    print('Failed to get data:', response.status_code)
else:
    #print('First 100 characters of data are')
    #print(response.text[:100])
    output.write(response.head())
output.close()
```
